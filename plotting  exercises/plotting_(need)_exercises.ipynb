### Exercise 1 
Can't tell if its me but the legend is not showing up right

```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label= 'Y values with errors')
plt.legend('lower right')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.show()
```

### Exercise 2

```python
import numpy as np
x = np.linspace(0,50,100)
y = np.exp(-x/10)*np.sin(x)

plt.plot(x,y,c = 'blue')
plt.grid(alpha = 0.25)
plt.xlabel('x', c = 'darkorange', fontsize = 15)
plt.ylabel('y', c = 'darkorange', fontsize = 15)
plt.title(r'$y=e^{ -x/10 } \sin(x)$', fontsize = 20 )
plt.ylim(top=1,bottom=-1)
```
