# 1 A simple plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='blue', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25) # This changes opacity of grid lines
plt.legend(loc='lower right')
```

## 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
# To plot another graph just add another line with the parameters u want
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
# use xerr and yerr for error bars, set to fixed values
# can error bars be variable? i.e percentages 
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

convert into np arrays first

```python
import numpy as np
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

can use numpy to input functions instead of just hard inputting data 

```python
x = [i for i in range (-10,10)]
x = np.array(x)
cos_x = np.cos(x)
y = 0.1*x**2 + x + 7 
plt.plot(x,y,c = 'red', ls = 'dotted', lw = 2, label = 'y = 0.1x^2 + x + 7')
plt.plot(x,cos_x, c = 'green', lw = 2, label = 'cos(x)')
plt.legend(loc = 'upper left')
```

## We need another axis!

```python
x = [i for i in range (-10,10)]
x = np.array(x)
cos_x = np.cos(x)
y = 0.1*x**2 + x + 7 

plt.plot(x,y,c = 'red', ls = 'dotted', lw = 2, label = 'y = 0.1x^2 + x + 7')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x,cos_x, c = 'green', lw = 2, label = 'cos(x)')
plt.legend(loc='lower right')                                 # For y-axis 2
```

### Things to note
1. There are two `legend()` calls, one for each axis
1. Use ` np.linspace()` to make the graph smoother
    * [link is here](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html)
    


```python
x = np.linspace(-10,10,200) # returns an array for (start,stop, number of points)
cos_x = np.cos(x)
y = 0.1*x**2 + x + 7 

plt.plot(x,y,c = 'red', lw = 2, label = 'y = 0.1x^2 + x + 7')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x,cos_x, c = 'green', lw = 2, label = 'cos(x)', ls = 'dashed')
plt.legend(loc='lower right')   

```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0] # all rows, column in question
anomaly = data[:, -1] # same here

plt.plot(date, anomaly, alpha=.5) #low alpha to soften plot colour
plt.ylim([-4, 4])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') 
# adds a horizontal line at y = 0 from first date to last date
```

## 4.2 Styles

Just type in `plt.style.use('style_name')` to use different styles and not configure the graph yourself. The link to the standard templates is [here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html)

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-4, 4])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.')
```

```python
plt.style.use('ggplot')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-4, 4])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.')
```
