```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
 
def coupled_differential_equations(y, t, k, b):
    susceptible, infected, recovered = y
    dsdt = -b*susceptible*infected
    drdt = k * infected
    didt = b*susceptible*infected-k*infected
    return [dsdt, drdt, didt]
 
y0 = [1000000.0, 1.0, 0.0]  
k = 5.0 
b = 0.2 
t = np.linspace(0, 10, 1000) # Time in days
 
solutions = odeint(coupled_differential_equations, y0, t, args=(k, b))
 
s_values = solutions[:, 0]
r_values = solutions[:, 1]
i_values = solutions[:, 2]
 
plt.plot(t, s_values, label='Susceptible')
plt.plot(t, r_values, label='Recovered')
plt.plot(t, i_values, label='Infected')
plt.xlabel('Time')
plt.ylabel('Position / Velocity')
plt.legend()
plt.title('SIR Model')
plt.show()
```
